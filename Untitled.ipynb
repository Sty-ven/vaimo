```python
import requests

import pandas as pd


df = pd.read_excel('location.xlsx')
```

```python

from datetime import date, datetime, timedelta

def datespan(startDate, endDate, delta=timedelta(days=1)):
    currentDate = startDate
    while currentDate < endDate:
        yield currentDate
        currentDate += delta
        
all_results = []  
def get_data():
    all_results = []
    for day in datespan(date(2007, 3, 30), date(2007, 7, 1),delta=timedelta(days=1)):
#         print(day)
        
        for name, lat, long in df[['Name', 'Latitude', 'Longitude']].values:
            url = f"https://api.sunrisesunset.io/json?lat={lat}&lng={long}&timezone=UTC&date={day}"
            try:
                r = requests.get(url)
                result_dict = r.json()['results']
                result_dict['date'] = day
                result_dict['name'] = name
                all_results.append(result_dict)
            except Exception as e:
                print(f"Day: {day}... Error: {e}")
                pass
    data = pd.json_normalize(all_results)
    data = data[['name', 'date', 'dawn', 'dusk', 'sunrise', 'sunset']]
    
    return data
    
```

```python
data = get_data()
```
